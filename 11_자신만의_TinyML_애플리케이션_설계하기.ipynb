## 11.1 설계 과정

- 임베디드 머신러닝 프로젝트의 위험 중 하나
  - 모델 훈련에는 며칠에서 몇 주의 시간이 걸릴 수도 있음
  - 새로운 임베디드 하드웨어 플랫폼을 구축하는 데는 많은 시간이 필요함
  - 실제로 무엇을 가동시키기 전에 시간이 부족해지는 것이 위험 요소


- 이 위험을 줄이는 가장 효과적인 방법
  - 계획, 연구, 실험을 통해 프로세스 초기에 가능한 한 많은 질문에 대한 답을 만드는 것임
  - 훈련 데이터나 신경망 아키텍처를 변경할 때마다 일주일 정도의 코딩과 훈련 시간은 쉽게 소요됨
  - 배포할 하드웨어를 변경하게 되면 소프트웨어 스택 전체에 파급 효과를 가져오게 되며 이전에 잘 작동하던 코드의 상당 부분을 새로 작성해야 함
  - 개발 프로세스 후반에 발생하는 변경 사항을 줄이려면 초반에 변수를 줄여서 시간을 절약해야 함


- 최종 애플리케이션 코딩에 앞서 중요한 질문에 답하기 위한 방법을 살펴보자


## 11.2 마이크로컨트롤러가 필요할까, 더 큰 장치가 필요할까?

- 첫째로 답해야 할 질문
  - 임베디드 시스템이 정말 필요한지?
  - 아니면 최소한 초기 프로토타입에서는 배터리 수명, 비용, 크기 등의 요구 사항을 완화할 수 있는지?


- 리눅스와 같은 완전한 최신 운영체제를 갖춘 시스템에서 프로그래밍 하는 것은 임베디드 세계에서 개발하는 것보다 훨씬 쉽고 빠름
  - 카메라나 센서와 같은 많은 주변 장치를 연결할 수 있는 저렴한 라즈베리파이 같은 기기는 테스크톱 수준의 시스템임
  - 계산량이 많은 신경망을 실행해야 한다면 엔비디아 젯슨 계열 보드가 작은 폼 팩터와 강력한 소프트웨어 스택을 갖추고 있어 적절함

- 하지만 젯슨 보드의 가장 큰 단점은 수 와트 수준의 전력 소모임
  - 에너지 저장 장치의 물리적 크기에 따라 배터리로 가동할 경우 몇 시간 또는 며칠 정도 작동하면 배터리가 방전될 것


- 서버 통신으로 인한 대기 시간이 문제되지 않는다면 강력한 클라우드 서버 여러 개를 사용해서 신경망 워크로드를 처리하고 클라이언트 장치가 인터페이스와 네트워크 통신을 처리하는 방식이 더 나을 수도 있음


- 어디에나 배포할 수 있는 범용성은 이상적인 개념이지만, 아이디어가 효과가 있는지 개념 증명(proof of concept)을 통해 판단하고 싶을 때는 쉽고 빠르게 실험할 수 있는 기기를 사용하여 프로토타입을 제작하는 것이 좋음
  - 임베디드 시스템을 통한 개발에는 까다로운 요소가 많기 때문에 실제 개발에 돌입하기 전에 애플리케이션의 실제 요구 사항을 많이 파악할 수록 프로젝트의 성공 확률은 높아짐


- 일례로 목장에서 기르는 양의 건강 상태를 모니터링하는 데 필요한 장치를 만든다고 상상해보자
  - 네트워크 연결 상태가 좋지 않은 환경에서 몇 주 혹은 몇 달 동안 실행할 수 있는 장치여야 하므로 임베디드 시스템이 필요함
  - 그러나 아직 어떤 모델을 쓸지, 어떤 센서가 필요할지, 어떤 동작을 할지 등의 정보를 알지 못하므로 프로그래밍하기 까다로운 장치를 사용하는 것은 별로 바람직하지 않음
  - 심지어 아직 훈련 데이터도 없음
  - 라즈베리파이를 써서 양을 모니터링하고, 충전을 위해 매일 밤 보드를 수거하는 방식을 사용하면서 목장 전체에서 사용할 수 있는 실외 와이파이 네트워크를 설치하면 양의 데이터를 수집할 수 있음
  - 실제로 이런 솔루션을 필요로 하는 고객은 없을 수 있지만, 이러한 환경을 설치해보면 새로운 모델, 센서, 폼 팩터를 실험하는 데 필요한 많은 질문에 대답할 수 있으며 임베디드 버전보다 훨씬 빠르게 진도를 나갈 수 있음


- 마이크로컨트롤러는 다른 하드웨어에 없는 확장성이라는 명확한 장점이 있음
  - 싸고 작으며 소모 전력도 매우 낮지만 이러한 장점은 실제로 확장이 필요할 때만 의미가 있음
  - 가능하면 목적이 명확하고 확장성이 필요한 것을 확신할 수 있을 때까지 확장성에 대한 고려는 미루자


## 11.3 무엇이 가능한지 이해하기

- 딥러닝으로 어떤 문제를 해결할 수 있는지 파악하는 것이 필요함
  - 신경망은 계획이나 명제 해결과 같은 작업을 자동으로 수행할 수 없음
  - 노이즈가 많고 혼란스러운 대량의 데이터를 수집하고 그 패턴을 파악하는 작업을 훨씬 잘 함


- 예를 들어 신경망은 양치기 개를 지휘해서 양떼를 목장 안으로 몰기는 어려움
- 하지만 양떼의 체온, 맥박, 가속도계와 같은 다양한 센서 데이터를 수집해서 양의 건강 상태를 예측하는 데는 적합함


- 사고력이 명시적으로 필요한 문제와 달리 사람이 거의 무의식적으로 수행하는 대부분의 판단은 딥러닝으로 다룰 가능성이 높음
  - 사람은 직관적으로 사물, 소리, 단어, 친구를 순간적으로 인식할 수 있는데, 이러한 종류의 작업은 딥러닝이 잘 수행하기에 알맞은 작업임

## 11.4 다른 사람의 발자취 따르기

- 선행 문헌 조사
  - 신경망 모델을 다양한 문제에 적용하기 위한 기존 연구 내용을 배우고 다른 사람이 이룬 업적을 바탕으로 이어서 작업하는 법을 익히면 많은 시간을 절약할 수 있음
  - 유사한 문제에 대해 사람들이 어떤 모델을 사용했는지, 그리고 사용할 수 있는 기존의 데이터셋이 존재하는지 여부를 확인
  - 머신러닝에서 가장 어려운 부분이 데이터 수집이기 때문에 활용할 수 있는 데이터셋을 확보하면 문제에 대한 접근 방식을 실험하는 데 도움이 되므로 매우 유용함


- 컴퓨터 분야 연구 논문을 검색할 수 있는 사이트
  - arXiv: 물리학, 수학, 컴퓨터 과학, 통계, 경제학 등의 분야 학술 논문을 위한 무료 배포 서비스이자 오픈 액세스 아카이브. 머신러닝 연구 논문이 활발히 올라오는 곳. (https://arxiv.org/)
  - ACM Digital Library (https://dl.acm.org/)
  - IEEE Xplore (https://ieeexplore.ieee.org/Xplore/home.jsp)
  - AI, ML 분야 국제 학회 (https://jackietseng.github.io/conference_call_for_paper/conferences-with-ccf.html)


- 예를 들어 딥러닝 기반의 기계식 베어링 결함 진단에 관심이 있다면, arxiv.org에서 deep learning bearing fault을 검색. 아래 논문에서 베어링 센서 데이터의 표준 공개 데이터셋이 있음을 알 수 있음
  - Machine Learning and Deep Learning Algorithms for Bearing Fault Diagnostics -- A Comprehensive Review https://arxiv.org/abs/1901.08247
  - Case Western Reserve University (CWRU) bearing dataset
  - 또한 문제점에 사용된 다양한 종류의 모델 아키텍처에 대한 개요와 장점, 비용, 전체 결과에 대한 평가도 확인할 수 있음



## 11.5 훈련할 모델 찾기

- 사용할 모델 아키텍처와 훈련 데이터에 대한 아이디어를 얻은 다음에는 자원 제약이 없는 훈련 환경에서 시간을 투자해 결과를 확인하는 것이 바람직함


- 텐서플로우 프레임워크를 기반으로 한다면 텐서플로우 튜토리얼이나 스크립트 예제를 찾아서 그대로 실행해본 다음 각 문제에 맞게 조정하는 방식을 취할 수 있음
  - 어떤 모델이 효과가 있을지 알아보려면 센서 데이터의 특성을 보고 튜토리얼에서 비슷한 모델을 활용하는 방법을 쓸 수 있음


- 예를 들어, 휠 베어링의 단일 채널 진동 데이터가 있다고 생각해보자
  - 이는 비교적 고주파수의 시계열 데이터로, 마이크로 수집한 오디오 데이터와 공통점이 많음
  - 그러면 우선 모든 베어링 데이터를 .wav 형식으로 변환한 다음 적절한 레이블을 붙인 후 표준 Speech Commands 데이터셋 대신 음성 훈련 프로세스에 이 데이터를 공급할 수 있음


- 제스처 인식 예제를 다른 가속도계 기반 분류 문제에 적용하거나 인체 감지 모델을 다른 머신 비전 애플리케이션에 적용할 수도 있음


- 문제에 따라서 예제 중 참조할 만한 예제가 없을 수도 있음. 그 경우 케라스를 사용해서 모델 아키텍처를 구축하는 튜토리얼을 참고하여 모델을 구축할 수 있음


- 기존 논문이나 github 사이트에서 공개한 모델 아키텍처를 참고하거나 활용할 수도 있음

## 11.6 데이터 관찰

- 머신러닝 연구의 초점은 대부분 새로운 아키텍처를 설계하는 데 있고 훈련 데이터셋에 대한 내용은 많지 않음
  - 일반적으로 미리 생성된 훈련 데이터셋이 제공되며 신모델이 다른 모델과 비교하여 점수를 얼마나 잘 얻을 수 있는지 경쟁하는 경우가 많기 때문임


- 많은 현실의 문제는 일반적으로 기존 데이터셋이 없으며 고정 데이터셋을 활용한 예측 점수가 아니라 최종 사용자에게 제공하는 경험이 더 중요하므로 프로젝트의 우선 순위가 달라짐
  - 모델 아키텍처에 대한 연구만큼 데이터의 수집, 탐색, 레이블 지정, 개선에 훨씬 더 많은 시간을 할애해야 함

#### 데이터 작업에 있어서 고려 사항

- 데이터를 관찰하고 데이터에 있는 문제를 해결하면 데이터 분석을 한층 더 발전시킬 수 있음
  - 이미지 데이터라면 로컬 시스템에 레이블 별로 정렬된 폴더로 이미지를 다운로드하고 하나씩 살펴보기
  - 오디오 파일이라면 이미지와 마찬가지로 폴더로 다운로드하고 선택한 파일을 들어보기


- 조금만 살펴봐도 생각지도 못한 실수나 이상한 점을 발견할 수도 있음
  - 예를 들어 재규어 자동차가 재규어(고양이과 동물)로 잘못 레이블링 되고, 오디오가 너무 희미하거나 단어의 일부가 잘린 녹음이 있을 수도 있음
  - 쉼표로 구분된 값(CSV) 파일의 숫자를 살펴보고 센서 값이 포화 값에 도달하거나 오버플로우가 발생했거나 감도가 너무 낮아서 대부분 데이터가 너무 작은 숫자 범위로 채워지는 것과 같은 문제가 있다는 것을 확인할 수도 있음





- 불균형 훈련 데이터셋 문제
  - 카테고리 분류 과제의 경우 학습 데이터셋에서 각 클래스의 빈도가 최종 예측 확률에 영향을 줌

  - 빠지기 쉬운 함정 중 하나는 네트워크 결과가 실제 확률을 나타내는 것이라고 착각하는 것임. 예를 들어 yes 음성에 대해 점수 0.5가 나오면 모델이 입력 음성이 yes일 확률을 50%로 예상했다는 의미임. 훈련 데이터에서 각 클래스의 비율이 출력값을 제어한다는 점을 감안하면 훨씬 더 복잡한 이야기가 되지만, 실제 확률을 이해하려면 애플리케이션의 실제 입력 분포에서 각 클래스의 사전 확률이 필요함


- 다른 예로 열 가지 종류의 새에 대한 이미지 분류기를 훈련시키는 과제를 상상해보자
  - 훈련한 모델을 남극에 배포했는데 앵무새라는 결과가 나타나면 의심스러울 것임. 모델을 아마존에 배포했는데 펭귄이 나타나도 마찬가지임.
  - 이러한 종류의 도메인 지식을 모델 훈련 과정에 적용하는 것은 매우 어려운데, 일반적으로 신경망이 각 클래스에 대해 동등하게 '주의를 기울이도록' 각 클래스에서 거의 동일한 수의 샘플을 원하기 때문임
  - 현실적으로는 모델 추론 시 후보정 프로세스를 도입해서 사전 지식을 기반으로 결과에 가중치를 부여하는 방법이 사용됨
  - 이를테면 남극의 예에서는 앵무새가 추론 결과가 되기 위한 임계값이 매우 높고, 펭귄의 임계값은 훨씬 낮게 하는 식임


## 11.7 오즈의 마법사 방법론

- 엔지니어링에서 가장 어려운 문제는 요구 사항이 무엇인지 결정하는 것임
  - 특히 머신러닝 모델을 개발하는 프로세스 때문에 실제로 문제에 효과적이지 않은 작업에 많은 시간과 리소스를 사용하는 일은 빈번하게 일어남


- 요구 사항을 정리하기 위한 방법으로 권장하는 방법 중 하나가 오즈의 마법사 방법임
  - 오즈의 마법사 방식은 실제로 구축하려는 시스템 모형을 만드는데, 소프트웨어를 통해 의사 결정을 내리는 대신 '커튼 뒤에 있는 사람'이 결정을 내림
  - 이를 통해 가정을 테스트 할 수 있으며 시간이 많이 걸리는 개발 주기를 거치기 전에 사양을 잘 테스트했는지 확인할 수 있음

- 사람들이 회의실에 있을 때 이를 감지하고 회의실에 사람이 없으면 조명이 어두워지는 센서를 설계한다고 하자
  - 오즈의 마법사 접근법에서는 인체 감지 모델을 실행하는 무선 마이크로컨트롤러를 만들어 배포하는 대신 회의실 근처 방에 사람을 한 명 두고 회의실 라이브 비디오를 볼 수 있게 하고, 조명을 제어하는 스위치로 프로토타입을 만듬
  - 이를 사용하는 과정에서 카메라가 회의실 전체를 볼 수 없어서 누군가가 있는데도 아무도 보이지 않아서 조명이 계속 꺼지거나, 누군가 회의실에 들어올 때 조명이 켜지는 지연 시간이 너무 긴 경우와 같은 사용성 문제를 빠르게 발견할 수 있음
  - 이 방법은 거의 모든 문제에 적용할 수 있으며 잘못된 가정을 기반으로 하는 머신러닝 모델에 시간과 에너지를 소비하지 않고도 제품에 대한 가정을 검증할 수 있음



## 11.8 데스크톱에서 먼저 작동시키기

- 오즈의 마법사 접근법은 프로토타입을 최대한 빨리 실행하는 한 가지 방법이지만, 모델 훈련으로 넘어간 다음에는 최대한 빨리 실험하고 반복하는 방법을 생각해야 함
  - 임베디드 플랫폼으로 모델을 내보내고 해당 모델을 빠르게 실행하려면 시간이 오래 걸릴 수 있음


- 빠르게 진도를 나가려면 센서 데이터를 데스크톱이 수집하게 하고 클라우드 시스템으로 처리하는 것이 좋을 수 있음
  - 이렇게 만든 시스템을 프로덕션 환경으로 바로 배포하면 상당한 에너지를 쓰는 솔루션이 되겠지만, 서버 통신의 대기 시간이 사용자 경험에 영향을 미치지 않는 한 이 방법으로 머신러닝 솔루션이 제품 설계 맥락에서 얼마나 잘 작동하는지에 대한 피드백을 얻을 수 있음


- 이 방식의 또 다른 장점은 센서 데이터를 한 번 기록한 다음 모델 평가를 위해 반복적으로 사용할 수 있다는 점
  - 모델에 영향이 큰 오류가 있고 이러한 오류가 정상적인 지표로 제대로 캡처되지 않을 때 특히 유용함
  - 사진 분류기가 아기를 개로 분류한다면 사용자에게 분노를 일으킬 수 있으므로 전반적으로 95%의 정확도를 가지고 있다 해도 모델의 재평가를 위해 이를 피하는 것이 좋음


- 데스크톱에서 모델을 실행하는 방법
  - 시작하기 가장 쉬운 방법은 센서 연동이 편리한 라즈베리파이와 같은 플랫폼을 사용하여 예제 데이터를 수집하고, 데스크톱 시스템(또는 클라우드 인스턴스)으로 복사하는 방법
  - 그런 다음 파이썬에서 표준 텐서플로우를 사용하여 모델을 오프라인 방식으로 학습하고 평가할 수 있음
  - 괜찮아 보이는 모델이 나오면 텐서플로우 모델을 텐서플로우 라이트 모델로 변환하는 것과 같은 단계를 수행
  - 데스크톱 텐서플로우 라이트 애플리케이션을 간단한 웹 API 뒤에 놓고 실제 환경에서의 작동 방식을 이해하기 위해 임베디드 폼 팩터의 장치에서 호출할 수 있음
