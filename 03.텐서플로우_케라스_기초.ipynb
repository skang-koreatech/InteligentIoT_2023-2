#### 3.4.3 사칙 연산

```python
import tensorflow as tf
import numpy as np

a = tf.constant(3)
b = tf.constant(2)

# 기본 연산
# 텐서 형태로 출력해보기
print(tf.add(a, b))  # 더하기: tf.add()
print(tf.subtract(a, b)) # 빼기: tf.subtract()

# 넘파이 배열 형태로 출력해보기
print(tf.multiply(a, b).numpy()) # 곱하기: tf.multiply()
print(tf.divide(a, b).numpy()) # 나누기: tf.divide()
```

#### 3.4.4 텐서 - 넘파이 배열 변환

```python
import tensorflow as tf
import numpy as np

c = tf.add(a, b).numpy() # a와 b를 더한 후 numpy 배열 형태로 변환
c_square = np.square(c, dtype=np.float32) # numpy 모듈에 있는 square 함수 적용
c_tensor = tf.convert_to_tensor(c_square) # numpy 배열을 다시 tensor로 변환

print('numpy array : %0.1f, applying square with numpy : %0.1f, convert_to_tensor : %0.1f' %(c, c_square, c_tensor))
print(c_tensor)
```

#### 3.4.5 논리 연산

```python
import tensorflow as tf

c1 = tf.constant([1, 3])
c2 = tf.constant([5, 7])
v1 = tf.constant([1, 9])
v2 = tf.constant([7, 7])

print("==== equal, not_equal ====")

# 등호 비교
print(tf.equal(c1, c2))
print(tf.not_equal(c1, c2))

print("==== greater, less =====")

# 대소 비교
print(tf.less(c1, c2))
print(tf.less_equal(c2, v2))
print(tf.greater(c1, v1))
print(tf.greater_equal(c1, v1))

print("==== and or xor not =====")

# and, or, xor, not
c3 = tf.constant(True)
c4 = tf.constant(False)
print(tf.logical_and(c3, c4))
print(tf.logical_or(c3, c4))
```

### 3.5 Keras API 이용 OR 게이트 구현

- 이전 AND, OR 게이트 구현에서는 가중치(weight와 bias를 포괄하여 가중치라고 표현)를 직접 설정해주고 이를 이용해 게이트 연산을 수행해 보았음
- 여기서는 단순 퍼셉트론과 학습 데이터를 이용해 가중치가 자동으로 학습되도록 하는 것을 살펴봄

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse

tf.random.set_seed(777)

# 데이터 준비
data = np.array([[0, 0], [1, 0], [0, 1], [1, 1]]) # OR 연산의 입력으로 사용될 x1, x2 데이터 (학습 데이터)
label = np.array([[0], [1], [1], [1]]) # OR 연산 결과 (정답/레이블 데이터)

# 모델 구성/정의
model = Sequential() # Sequential API 이용 모델 객체 생성
model.add(Dense(1, input_shape = (2, ), activation = 'linear')) # 단층 퍼셉트론 구성 - 모델에 노드 한 개짜리 Dense 층 추가. 노드의 입력 데이터 2개

# 모델 준비(컴파일)
model.compile(optimizer = SGD(), loss = mse, metrics = ['acc'])
#model.compile(optimizer='sgd', loss='mse', metrics=['acc'])

# 모델 학습/훈련
model.fit(data, label, epochs = 500)
```

```python
model.get_weights()
```

#### 3.5.1 주요 코드 설명

model.add(Dense(1, input_shape = (2, ), activation = 'linear'))
- model.add() 함수는 신경망 모델에 하나의 층을 추가하는데 사용

Dense(1, input_shape = (2, ), activation = 'linear')
- Dense(): Dense 층을 생성
  - 밀집층(Dense layer), 완전 연결층(Fully-connected layer), 다층 퍼셉트론(MLP) 등 여러 가지 표현으로 불림 
  - 첫 번째 파라미터 1 : 노드(퍼셉트론) 개수 (출력 1개)
  - 두 번째 파라미터 input_shape=(2, ) : 노드의 입력 데이터의 형태를 말함. 2개의 특성을 가지는 1차원의 데이터라는 의미 
  - 세 번째 파라미터 activation='linear' : 활성화 함수를 설정. 'linear'는 항등 함수 (디폴트 값, 입력뉴런과 가중치로 계산된 결과값이 그대로 출력으로 나옴. 선형 회귀 모델에서 사용)


model.compile(optimizer = SGD(), loss = mse, metrics = ['acc'])
- 모델 훈련/학습에 필요한 파라미터 설정
  - optimizer : 최적화 방법 (경사하강법과 같이 최적의 가중치를 찾아가는 방법들이 여러 가지가 있는데 그 중 어떤 것을 사용하지 설정하는 것). 대표적으로 SGD(), RMSProp(), Adam(), NAdam() 등이 있음. 'sgd', 'rmsprop', 'adam', 'nadam' 같이 문자열로 지정할 수도 있음. SGD는 Stochastic Gradient Descent의 약자로 확률적 경사하강법이라고 하며 전체 학습 데이터가 아닌 일부 데이터에 대해 loss를 계산하여 가중치를 업데이트하여 계산이 빠름. tf.keras.optimizers 모듈에 정의됨
  - loss : 손실(loss) 함수. mse는 Mean Squared Error(regression에서 사용). 이외에 binary_crossentropy (binary classification에서 사용), categorical_crossentropy (multi-class classification에서 사용) 등이 있음. tf.keras.losses 모듈에 정의됨
  - metrics : 학습된 모델의 성능을 평가하기 위한 평가 지표. 리스트로 전달하며, acc는 accuracy(정확도)를 의미


model.fit(data, label, epochs = 500)
- 모델 학습 수행. 
  - 첫 번째 파라미터 : 학습용 입력/특성 데이터
  - 두 번째 파라미터 : 정답/레이블 데이터
  - epochs : 학습 데이터 반복 횟수 
  - 이 외에 batch_size(한 번의 학습에 사용할 데이터의 개수), validata_data(검증용 데이터) 파라미터를 설정할 수 있음 


model.get_weights()
- 학습된 모델의 가중치와 편향을 반환
- 출력 결과에서 2개의 가중치 (0.494, 0.375), 1개의 편향 (0.327) 값을 확인할 수 있음

### 3.6 XOR 게이트 구현

위의 예제와 유사하게 단순 퍼셉트론을 이용하여 XOR 게이트의 가중치를 학습할 수 있는지 살펴보자


```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse

tf.random.set_seed(777)

# 데이터 준비
data = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
label = np.array([[0], [1], [1], [0]])

# 모델 구성/정의
model = Sequential()
model.add(Dense(1, input_shape = (2, ), activation = 'linear'))

# 모델 준비/컴파일
model.compile(optimizer = SGD(), loss = mse, metrics = ['acc'])

# 모델 학습
model.fit(data, label, epochs = 500)
```

위 결과에서 확인할 수 있는 것
- 100 epoch 정도부터 평가지표인 accuracy가 0.5로 동일하게 유지되고, loss도 0.25정도로 더 감소하지 않음
- 학습이 이루어지지 않고 있음을 알 수 있음

이전에 확인했던 것처럼 단순 퍼셉트론으로는 XOR 게이트를 표현할 수 없다는 것을 알 수 있음

#### 3.6.1 다층 퍼셉트론으로 XOR 게이트 구현

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse

tf.random.set_seed(777)

# 데이터 준비
data = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
label = np.array([[0], [1], [1], [0]])

# 모델 구성/정의
model = Sequential()
model.add(Dense(32, input_shape = (2, ), activation = 'relu'))
model.add(Dense(1, activation = 'linear'))

# 모델 준비/컴파일
model.compile(optimizer = SGD(), loss = mse, metrics = ['acc'])

# 모델 학습
model.fit(data, label, epochs = 300)
```

- randint 함수를 이용해 60000 미만 임의의 정수 3개를 선택하여 이를 인덱스로 갖는 x_train, y_train 데이터를 출력

#### 3.7.2 데이터 전처리


```python
# 데이터 전처리 - 모델 학습을 위한 입력 데이터를 만들기 위한 과정

num_x_train = x_train.shape[0]
num_x_val = x_val.shape[0]
num_x_test = x_test.shape[0]

# 모델 입력을 위해 데이터 형태(shape)를 변경 
x_train = (x_train.reshape((num_x_train, 28*28))) / 255
x_val = (x_val.reshape((num_x_val, 28*28))) / 255
x_test = (x_test.reshape((num_x_test, 28*28))) / 255

print(x_train.shape) 
```

- 데이터 shape 변경
  - reshape 함수 사용: x 데이터는 28x28 이미지이므로 2차원 배열 형태인데, 784개 원소를 갖는 1차원 배열 형태로 만들어줌

- 데이터 스케일링
  - 신경망은 입력 데이터의 범위와 스케일에 매우 민감하기 때문에 적절한 정규화(Normalization)/표준화(Standardization) 필요
  - 위 코드에서는 255로 나누어 줌. 이것은 각 픽셀이 가질 수 있는 최대값으로 나누어 준 것으로 MinMax Normalization 적용한 것임. 


- MinMax Normalization
  - 데이터 값을 0에서 1 사이 값이 되도록 변경하는 것
  - 각 데이터 값에서 데이터 최소값을 빼고 이를 데이터 최대값과 최소값의 차로 나누어 줌
  - X를 [x_1, x_2, x_3, ..., x_n]이라 할 때 x_i-new = (x_i - min(X)) / (max(X) - min(X))
  - MinMax scaling이라고도 함
 


- 표준화(Standardization)
  - 전체 데이터의 평균을 0, 표준편차를 1이 되도록 값을 변경하는 방법
  - 각 데이터 값에서 데이터의 평균을 빼고 이를 표준편차로 나누어 줌
  - Z-score Normalization(Z-점수 정규화)이라고도 함

```python
# 모델 입력을 위한 레이블 데이터 전처리 

from tensorflow.keras.utils import to_categorical

# 각 데이터의 레이블을 범주형 형태로 변경
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)

print(y_train)
print(y_train.shape)
```

- y 데이터, 즉 레이블 데이터(정답 데이터)는 0에서 9까지 정수로 표현됨
- 분류 문제에서는 이를 **범주형 데이터**로 변환 필요
- 즉, 총 10개의 범주(0-9) 중 어느 범주(클래스)에 속하는지 표현하는 방식
  - 10개의 원소를 갖는 배열이 되며 해당 숫자에 해당하는 원소만 1이고 나머지 원소는 0으로 표현됨
  - 이런 변환을 one-hot encoding이라고 함

#### 3.7.3 모델 구성/정의

- 아래 예제에서는 3개 Dense 층으로 구성
- 첫 번째 층에서는 입력 데이터 형태를 명시
- 마지막 층은 0-9까지 숫자 중 어느 것에 해당하는지 결과를 내야 하므로 출력층의 노드 수는 10개로 하고 활성화 함수는 Softmax 함수 사용

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# 첫 번재 층에서는 입력 데이터의 형태를 반드시 명시. input_shape 파라미터
# 784차원의 데이터를 입력으로 받고, 64개의 출력을 가지는 첫 번째 Dense 층
model.add(Dense(64, activation='relu', input_shape=(784, )))

model.add(Dense(32, activation='relu'))     # 32개의 출력을 가지는 Dense 층

model.add(Dense(10, activation='softmax'))  # 10개의 출력을 가지는 Dense 층 - 출력층
```

#### 3.7.4 모델 컴파일

- 옵티마이저, 손실 함수, 평가 지표 설정
  - Adam optimizer 사용 (기본 학습률: 0.001)
  - 참고 사이트: https://keras.io/ko/optimizers/
  - 다중 분류 문제에서 손실 함수는 categorical_crossentropy 함수 사용
  - 평가 지표는 정확도 사용


```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

#### 3.7.5 모델 학습

- 모델 학습을 위한 x_train, y_train 데이터 입력
- epochs: 학습 반복 횟수, 전체 데이터를 30회 반복
- batch_size: 128개의 데이터를 하나의 배치로 사용
  - 총 42000개 데이터이므로 329 배치 (42000 / 128 = 328 나머지 16)
- validation_data: 검증용 데이터
  - 위에서 나누어놓은 검증 데이터 x_val, y_val 사용

```python
# 모델 학습하기

history = model.fit(x_train, y_train, 
                    epochs=30, batch_size=128, validation_data=(x_val, y_val))
```

- 출력 결과를 보면 loss, acc, val_loss, val_acc 확인 가능
  - loss, acc는 학습 데이터에 대한 손실과 정확도
  - val_loss, val_acc는 검증 데이터에 대한 손실과 정확도

- model.fit() 함수 반환: History 객체
  - 학습/검증 데이터에 대한 손실, 평가지표 값을 딕셔너리 형태로 저장

```python
# history를 통해 확인해볼 수 있는 값 출력하기

history.history.keys()
```

```python
# 학습 결과 그려보기

import matplotlib.pyplot as plt

his_dict = history.history

loss = his_dict['loss']
val_loss = his_dict['val_loss']

epochs = range(1, len(loss) + 1)

fig = plt.figure(figsize=(10, 5))

# 학습 및 검증 손실 그리기
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color='blue', label='train_loss')
ax1.plot(epochs, val_loss, color='orange', label='val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc = his_dict['acc']
val_acc = his_dict['val_acc']

# 학습 및 검증 정확도 그리기
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, acc, color='blue', label='train_acc')
ax2.plot(epochs, val_acc, color='orange', label='val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('acc')
ax2.legend()

plt.show()
```

- 에폭 당 손실과 정확도를 그래프로 표현
- 에폭이 진행됨에 따라 학습 데이터와 검증 데이터 사이에 손실과 정확도 그래프가 점차 벌어짐을 볼 수 있음
  - 과대적합 되고 있다고 할 수 있음
- 과대적합이 발생하면 데이터 특성, 모델 구조 등을 수정하고 재학습하거나 과대적합이 일어나기 전까지 모델을 사용할 수도 있음

---

##### 과대적합(Overfitting)
- 모델이 학습 데이터에서는 좋은 성능을 보이지만 새로운 데이터에 대해서는 좋은 성능을 보이지 못하는 결과를 의미
  - 모델이 학습 데이터에만 맞추어져서 일반화하지 못한 상태
  - 모델이 너무 복잡할 경우 발생할 수 있음

- 해결 방법
  - 학습 데이터를 더 다양하게 많이 수집하여 학습에 사용
  - 정규화(Regularization)사용
  - 데이터에서 이상치(outlier) 제거


##### 과소적합(Underfitting)
- 모델이 학습 데이터를 충분히 학습하지 않아 모든 측면에서 성능이 나쁜 것을 의미
  - 모델의 성능이 개선될 여지가 많이 남은 상태
  - 모델이 너무 단순한 경우에 발생할 수 있음

- 해결 방법
  - 학습 데이터를 더 다양하게 많이 수집하여 학습에 사용
  - 더 복잡한 모델을 사용
  - 모델을 충분히 학습시킴

#### 3.7.9 용어 정리

##### Confusion matrix (혼동행렬)

Binary class(이진 분류)의 예

|  |  |  |  |
|---|---|---|---|
| | | 예측된 정답| 
| | | True | False |
| 실제 정답 | True | TP | FN | 
|  | False | FP | TN | 

- TP (True Positive): True 예측이 실제 True 정답을 맞춘 개수
- TN (True Negative): False 예측이 실제 False 정답을 맞춘 개수
- FN (False Negative): 실제 True이나 False로 잘못 예측한 개수
- FP (False Positive): 실제 False이나 True로 잘못 예측한 개수

---

##### Accuracy (정확도)
전체 데이터 중에서 실제 데이터의 정답과 모델이 에측한 정답이 같은 비율


Accuracy = (TP + TN) / (TP + TN + FP + FN)

- 정확도의 문제점
  - 데이터가 불균형할 때(클래스 별 데이터 수가 차이가 날 때) 평가지표로 적절하지 못 함
  - 코로나 음성 케이스가 990개 양성 케이스가 10개 일 때, 모두 음성이라고 해버리면 정확도는 99%로 매우 높음. 하지만 소수이지만 양성 케이스를 잘 분류하는 것이 중요하므로 정확도가 아닌 다른 지표를 사용하는 것이 필요

---

##### Precision (정밀도) / Recall (재현율)
Precision: 모델이 True로 예측한 정답 중 실제 정답이 True인 비율


Precision = TP / (TP + FP)


Recall: 실제 정답이 True인 것 중에서 모델이 True로 예측한 비율. Sensitivity (민감도)라고도 함


Recall = TP / (TP + FN)


- Precision과 Recall은 서로 trade-off 관계에 있음. precision을 높이면 recall아 내려가고 recall을 높이면 precision이 내려가는 경향이 있음

---

##### F1-score
Precision과 recall의 조화평균으로 계산됨

F1 = 2 x (precision x recall) / (precision + recall)
